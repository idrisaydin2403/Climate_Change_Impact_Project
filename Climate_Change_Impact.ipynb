```python
from sklearn.model_selection import train_test_split
```

```python
df = df.drop(columns = ['Country', 'Region', 'Adaptation_Strategies'])
```

```python
df['Crop_Type'] = df['Crop_Type'].map({
    'Corn': 0,
    'Wheat': 1,
    'Coffee': 2,
    'Sugarcane': 3,
    'Fruits': 4,
    'Rice': 5,
    'Barley': 6,
    'Vegetables': 7,
    'Soybeans': 8,
    'Cotton': 9
})
```

```python
df
```

```python
X = df.drop(columns = ['Crop_Yield_MT_per_HA'])
y = df['Crop_Yield_MT_per_HA']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 7)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Gradient Boosting Regressor': GradientBoostingRegressor(),
    'SVR': SVR()
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
model_performances = []
```

```python
for name, model in models.items():
    
    model.fit(X_train, y_train)
    
    prediction = model.predict(X_test)
    
    mae = mean_absolute_error(prediction, y_test)
    mse = mean_squared_error(prediction, y_test)
    r2 = r2_score(prediction, y_test)
    
    model_performances.append({
        'Name': name,
        'Mean Absolute Error': mae,
        'Mean Squared Error': mse,
        'R^2 Score': r2,
        'Model Score': model.score(X_test, y_test)
    })   
```

```python
model_performances_df = pd.DataFrame(model_performances)
```
